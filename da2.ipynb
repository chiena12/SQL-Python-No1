## Câu 6 : Find content added in the last 5 years

SQL 

SELECT  *,date_added
FROM    netflix_titles

## Câu 10: Find each year and the average numbers of content release by India on netflix and return top 5 year with highest avg content release

SQL 

SELECT  YEAR(CAST(date_added AS DATE)) AS Year,
        COUNT(*),
        ROUND(COUNT(*)*1.0/(SELECT COUNT(*) FROM netflix_titles WHERE country LIKE '%India%') *100,2) as avg_content_per_year
FROM netflix_titles
WHERE country LIKE '%India%'
GROUP BY  Year
ORDER BY avg_content_per_year
LIMIT 5



```python
df.head(1)
```

```python
df_india=df[df['country'].str.contains('India',na=False)]
df_india['year']=pd.to_datetime(df_india['date_added']).dt.year
df_result=df_india.groupby('year').show_id.count().sort_values(ascending=False).reset_index().rename(columns={'show_id':'avg_content'})
df_result['avg_content_per_year']=df_result['avg_content']/len(df_india)*100
df_result
```

## Câu 14 : Find the top 10 actors who have appeared in the highest number of movies produced in [India]

SQL 

SELECT  TOP 10
        TRIM(s.value) AS cast_name,
        COUNT(*) AS number_of_movies
FROM netflix_titles
CROSS APPLY SPRING_SPLIT(cast,',') AS s
WHERE country LIKE '%India%'
GROUP BY cast_name
ORDER BY number_of_movies DESC


    

```python
df.head(1)
```

```python
df14=df.copy()
df14['cast']=df14['cast'].str.split(',')
df14_exploded=df14.explode('cast')
df14_exploded['cast']=df14_exploded['cast'].str.strip()
df_cast=df14_exploded[(df14_exploded['country'].str.contains('India',na=False)) & (df14_exploded['type']=='Movie')]
df_cast['cast'].value_counts(ascending=False).reset_index().rename(columns={'count':'number_of_movies'}).head(10)
```

## Câu 15: Categorize the content based on the presence of the keywords 'kill' and 'violence' in the description field. Label content containing these keywords as 'Bad' and all other content as 'Good'. Count how many items fall into each category.

SQL 

SELECT  CASE 
            WHEN description LIKE '%kill%' OR description LIKE '%violence%' THEN 'Bad' ELSE 'Good' END AS review,
        COUNT(*) AS number_of_contents
FROM netflix_titles        
GROUP BY review


```python
df.head(1)
```

```python
import numpy as np
df_result=df.assign(review = lambda df: np.where(df['description'].str.contains('kill|violence',na=False),'Bad','Good'))
df_result['review'].value_counts().reset_index()
```
